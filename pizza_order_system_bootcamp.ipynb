```python
class Pizza:
 def get_description(self):
   return self.__class__.__name__
 def get_cost(self):
   return self.__class__.cost

#Subclasses


class Classic(Pizza):
  cost=115.0

  def __init__(self):
     self.description="The pizza is Classic pizza and ingredients are Pepperoni,cheese,mushroom,cheese,tomato\n"
     print(self.description)

class Margherita(Pizza):
  cost=90.0

  def __init__(self):
     self.description="The pizza is Margherita pizza and ingredients are Tomato and cheese\n"
     print(self.description)

class TurkPizza(Pizza):
  cost=110.0

  def __init__(self):
     self.description="The pizza is Turk pizza and ingredients areMeat,cheese,onion,pepper\n"
     print(self.description)

class PlainPizza(Pizza):
  cost=110.0

  def __init__(self):
     self.description="The pizza is Plain pizza and ingredients are Tuna,onion,cheese\n"
     print(self.description)


#Superclass

class Decorator(Pizza):
 def __init__(self, topping):
        self.component = topping

 def get_cost(self):
        return self.component.get_cost() + \
          Pizza.get_cost(self)

 def get_description(self):
        return self.component.get_description() + \
          ' ;' + Pizza.get_description(self)

        


class Olives(Decorator):
  cost = 3.0

  def __init__(self, sauce):
    Decorator.__init__(self, sauce)

class Mushrooms(Decorator):
  cost = 5.0

  def __init__(self, sauce):
    Decorator.__init__(self, sauce)    


class GoatCheese(Decorator):
  cost = 2.5

  def __init__(self, sauce):
    Decorator.__init__(self, sauce)    



class Meat(Decorator):
  cost = 8.0

  def __init__(self, sauce):
    Decorator.__init__(self, sauce)    


class Onions(Decorator):
  cost = 3.0

  def __init__(self, sauce):
    Decorator.__init__(self, sauce)    

class Corn(Decorator):
  cost = 1.5

  def __init__(self, sauce):
    Decorator.__init__(self, sauce) 

def isValid(credit_card):
   """
    Check if a credit card number is valid using the Luhn algorithm.
    """
   credit_card = str(credit_card)
    # Reverse the card number and convert each character to an integer
   digits = [int(x) for x in credit_card[::-1]]
    # Double the value of every second digit
   doubled_digits = []
   for i, digit in enumerate(digits):
        if i % 2 == 1:
            doubled_digits.append(digit * 2)
        else:
            doubled_digits.append(digit)
    # Add the digits of the doubled values and the undoubled digits
   total = 0
   for digit in doubled_digits:
        if digit > 9:
            total += digit - 9
        else:
            total += digit
    # If the total is divisible by 10, the card number is valid
   return total % 10 == 0




```
